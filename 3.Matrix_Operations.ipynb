```python
import numpy as np
```

```python
# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)
```

```python
# matrix addition
addition = A + B
print("Addition:\n", addition)
```

```python
# matrix substraction
subtraction = A - B
print("Subtraction:\n", subtraction)
```

```python
# matrix multiplication
multiplication = np.dot(A, B)
print("Multiplication:\n", multiplication)
```

```python
# transpose
transpose_A = A.T
print("Transpose of A:\n", transpose_A)
```

```python
# determinant
determinant_A = np.linalg.det(A)
print("Determinant of A:", determinant_A)
```
