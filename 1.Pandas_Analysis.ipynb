```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "size": [800, 1000, 1200, 1500, 1800],
    "rooms": [2, 2, 3, 3, 4],
    "price": [2000000, 2500000, 3000000, 4000000, 5000000]
}

df = pd.DataFrame(data)

df.to_csv("data.csv", index=False)
```

```python
df = pd.read_csv("data.csv")   # change filename if needed
df.head()
```

```python
# calculate average
avg_price = df['price'].mean()
avg_price
```

```python
# visualizations
```

```python
# bar chart
df['price'].plot(kind='bar', figsize=(8,4))
plt.title("Bar Chart of House Prices")
plt.xlabel("Index")
plt.ylabel("Price")
plt.show()
```

```python
# scatter plot
plt.scatter(df['size'], df['price'])
plt.title("Size vs Price")
plt.xlabel("Size (sqft)")
plt.ylabel("Price")
plt.show()
```

```python
# heatmap
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
### Insights and Observations
- The heatmap shows a strong positive correlation between house size and price.
- The number of rooms is also positively correlated with price.
- This indicates that as size and rooms increase, house prices tend to increase.
```
